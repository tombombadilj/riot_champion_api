```python
def correlation_p_value(data):
    a = [(x, y) for x in list(data) for y in list(data) if x != y]
    compare = list()
    for (x,y) in a:
        if (x, y) not in compare and (y, x) not in compare:
            compare.append((x, y))
    for (x, y) in compare:
        print x + " and " + y
        print scipy.stats.pearsonr(data[x], data[y])
        print "="*50
```

```python
import nltk, re, pprint
from nltk import word_tokenize
```

```python
# for tags and keywords without sentences
def cleaning_text_data(data):
    data_token = data.apply(lambda x:nltk.Text(word_tokenize(x)))
    data_lower = data_token.apply(lambda x: map(lambda y: y.lower(),x))
    stopwords = nltk.corpus.stopwords.words('english')
    data_stopwords = data_lower.apply(lambda x: [i for i in x if i not in stopwords])
    #stemming
    porter = nltk.PorterStemmer()
    data_stem = data_stopwords.apply(lambda x: [porter.stem(i) for i in x])
    #lemmatize
    wnl = nltk.WordNetLemmatizer()
    data_lemmatize = data_stem.apply(lambda x: [wnl.lemmatize(i) for i in x])
    # remove non-alphabet characters
    data_alpha = data_lemmatize.apply(lambda x: filter(lambda y: y.isalpha(), x))
    data_final = data_alpha.apply(lambda x: " ".join(x))
    return data_final
```

```python
tf_cleaned = cleaning_text_data(tf)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
class topic_grouping(object):
    def __init__(self, data, no_components):
        self.data = data
        self.no_components = no_components
    
    def process(self):
        #ignore terms that only appeared in 1 document and remove stop_words
        vectorizer = TfidfVectorizer(min_df = 2, stop_words='english')
        vectorizer.fit(self.data)
        self.tf = vectorizer.transform(self.data)
        self.name_features = vectorizer.get_feature_names()
        
    def labels(self):
        nmf = NMF(n_components=self.no_components, random_state=42)
        nmf.fit(self.tf)
        self.features = nmf.transform(self.tf)
        index_list = list()
        index_list.append(self.name_features)
        self.components_df = pd.DataFrame(nmf.components_).transpose().set_index(index_list)
    
    def grouping_check(self, no, no_top):
        return self.components_df[no].sort_values(0, ascending = False)[0:no_top]
```

```python
saving_data = dict()
for i in range(2,20):
    name = "n_%d" % i 
    saving_data[name] = topic_grouping(tf_cleaned, i)
    saving_data[name].process()
    saving_data[name].labels()
```

Working with Allytips - dimention reduction with Truncated SVD

```python
allytips = pd.read_sql_query("""SELECT id, full_allytips
                                FROM allytips""", conn)
```

```python
allytips_text = allytips.set_index("id")
```

```python
allytips_cleaned = cleaning_text_data(allytips_text["full_allytips"])
```

```python
# building a scikit-learn pipeline to reduce dimention of allytips using truncatedSVD (PCA for sparse data)
# for data preprocessed using TFIDF
from sklearn.pipeline import Pipeline
from sklearn.decomposition import TruncatedSVD
```

```python
class dim_red_SVD(object):
    def __init__(self, data, no_comp):
        self.data = data
        self.no_comp = no_comp
        
    def pipeline(self):
        vectorizer = TfidfVectorizer(min_df = 2, stop_words='english')
        TSVD = TruncatedSVD(random_state=42, n_components = self.no_comp)
        steps = [("tfidf", vectorizer),("truncated SVD", TSVD)]
        pipeline = Pipeline(steps)
        self.new_features = pipeline.fit_transform(self.data)
        self.sum_variance = pipeline.named_steps["truncated SVD"].explained_variance_ratio_.sum()
```

```python
data_SVD = dict()
ratio_sum_SVD = dict()
```

```python
for i in range(2,200):
    a = dim_red_SVD(allytips_cleaned, i)
    a.pipeline()
    data_SVD[a.no_comp] = a.new_features
    ratio_sum_SVD[a.no_comp] = a.sum_variance
```

```python
plt.scatter(x = ratio_sum_SVD.keys(), y = ratio_sum_SVD.values())
plt.xlabel("No. of components")
plt.ylabel("Sum of Explained Variance")
plt.title("No of components and sum of explained variance")
plt.show()
```
